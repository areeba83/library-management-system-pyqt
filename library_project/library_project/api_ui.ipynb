```python
# -------------------------------
# STEP 1: Imports
# -------------------------------
import requests
import ipywidgets as widgets
from IPython.display import display, clear_output

# -------------------------------
# STEP 2: Function to get quote
# -------------------------------
def get_quote():
    url = "https://quotes-by-api-ninjas.p.rapidapi.com/v1/quotes"

    headers = {
        "X-RapidAPI-Key": "ED06ECB13EMSH2BD06E7B56F4167P1F1B18JSN96879E269843",  # <-- tumhari key
        "X-RapidAPI-Host": "quotes-by-api-ninjas.p.rapidapi.com"
    }

    try:
        response = requests.get(url, headers=headers, timeout=5)
        data = response.json()
        quote_text = f"{data[0]['quote']}\n\n— {data[0]['author']}"
        return quote_text
    except Exception as e:
        return f"Error fetching quote:\n{e}"

# -------------------------------
# STEP 3: UI Components
# -------------------------------
button = widgets.Button(
    description="Get Quote",
    button_style="success",  # green button
    layout=widgets.Layout(width='200px', height='40px')
)

output = widgets.Output(layout=widgets.Layout(border='1px solid black', width='400px', height='120px'))

# -------------------------------
# STEP 4: Button Logic
# -------------------------------
def on_button_click(b):
    with output:
        clear_output(wait=True)
        print(get_quote())

button.on_click(on_button_click)

# -------------------------------
# STEP 5: Display UI
# -------------------------------
display(button, output)
```
